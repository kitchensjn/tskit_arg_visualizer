```python
import msprime
import random
import tskit_arg_visualizer
```

```python
# Generate a random tree sequence with record_full_arg=True so that you get marked recombination nodes
ts_rs = random.randint(0,10000) 
print(ts_rs)  
ts = msprime.sim_ancestry(
    samples=3,
    recombination_rate=1e-8,
    sequence_length=1_000,
    population_size=10_000,
    record_full_arg=True,
    random_seed=3626
)

print(ts.draw_svg(size=(1000,300), time_scale="rank", x_scale="treewise"))
```

```python
d3arg = tskit_arg_visualizer.D3ARG.from_ts(ts=ts)

d3arg.set_node_styles([
    {
        "id":10,
        "stroke_width":3,
        "symbol":"d3.symbolStar"
    },
])

d3arg.draw(
    width=550,
    height=500,
    edge_type="ortho"
)

d3arg.draw_node(
    node=10,
    degree=(1,2),
    y_axis_labels=False,
    tree_highlighting=False
)
```
